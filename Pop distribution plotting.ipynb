```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
import scipy as sp
from scipy import stats
#flipping a coin
n = 10
p =.5
k = np.arange(0,11)
bin_mass = stats.binom.pmf(k,n,p)
```

```python
plt.plot(k, bin_mass,'o-')
plt.title("Binominal: n=%i, p=%.2f" % (n, p), fontsize=15)
plt.xlabel('x: how many times a coin shows head(or tail)')
plt.ylabel('p: the probability of each times ')
```

```python
size = 1000
bin_random_trial = stats.binom.rvs(n=n,p=p,size=size)
plt.hist(bin_random_trial, bins=10, normed=True)
```

```python
# Poisson Distritbuion
# Let's say, people catch a cold twice a yaer
# ramda is 2, time is 1 (a year)
# A person would catch a cold 6 times less than .01: 1%
mu = 2 # mu = lmabda*time
k = np.arange(0,11)
pois = stats.poisson.pmf(k, mu)
```

```python
plt.plot(pois,'o-')
plt.title("Poisson: mu=%.2f" % (mu), fontsize=15)
plt.xlabel('x: how many times a specific event will occur during a period')
plt.ylabel('p: the probability for each')
```

```python
#Beta distirbution 
#The shape of distribution is depending on values of alpha and beta
a,b=.5,.5
x = np.arange(0,1,.01)
beta = stats.beta.pdf(x,a,b)
plt.plot(x,beta)
```

```python
#Exponential distribution
x=np.arange(0,15,.01)
lamb=.5
y=lambda x,l: l*np.exp(-l*x)
plt.plot(x,y(x,lamb))
plt.title("Exponential: lambda(aka. an error rate)=%.2f" % (lamb), fontsize=15)
plt.xlabel('x')
plt.ylabel('p: Th probability density')
```

```python
scale = (1/(lamb)**2)/2 #standard error (aka. ER) # SE=sqrt(V(E)) # V(E)=1/lambda**2
exp_random = stats.expon.rvs(scale=scale, size =1000)
plt.hist(exp_random, bins=20, normed=True)
```

```python
print('mean:%.3f, nearly equals to theoritical mean %.3f' %(exp_random.mean(), 1/lamb))
print('SD:%.3f, nearly eaquals to theoritical SE(or scale) %.3f' %(np.std(exp_random, ddof=1),scale))
```

```python
significant_level=.05
print('Null hypothesis: flipping a coin 10 times, and it lands heads up more than 9 times')
print('The result:H0 is {}'.format(bin_mass[9]+bin_mass[10]>significant_level))
```

```python
#covariable
def cov(x,y):
    if len(x) != len(y): return
    n = len(x)
    mean_x = np.array(x).mean()
    mean_y = np.array(y).mean()
    #xy = [(x[i]-mean_x)*(y[i]-mean_y) for i in range(n)]
    #return (np.array(xy).sum())/(n-1)
    xy = [x[i]*y[i] for i in range(n)] 
    return (np.array(xy).sum() - n*mean_x * mean_y) / (n-1)
x = [2.1, 2.5, 3.6, 4.0]
y = [8, 10, 12, 14]
print ('Covariance : ' + str(cov(x,y)) ) 
```

```python
df = pd.DataFrame(dict(age=y, address_range=x));df.cov()
```

```python
path = 'http://archive.ics.uci.edu/ml/machine-learning-databases/auto-mpg/auto-mpg.data'
mpg_data = pd.read_csv(path, delim_whitespace=True, header=None,
            names = ['mpg', 'cylinders', 'displacement','horsepower',
            'weight', 'acceleration', 'model_year', 'origin', 'name'],
            na_values='?')
```

```python
mpg_data[['mpg', 'weight']].corr()
```

```python
#or simply
mpg_data['mpg'].corr(mpg_data['weight'])
## mpg and weight have negative correlation
```

```python
mpg_data.drop(['model_year', 'origin'], axis=1).corr(method='spearman').style.format("{:.2}").background_gradient(cmap=plt.get_cmap('coolwarm'), axis=1)
```

```python
# plot correlated values
plt.rcParams['figure.figsize'] = [16, 6]
fig, ax = plt.subplots(nrows=1, ncols=3)
ax=ax.flatten()
cols = ['weight', 'horsepower', 'acceleration']
colors=['#415952', '#f35134', '#243AB5']
j=0
for i in ax:
    if j==0:
        i.set_ylabel('MPG')
    i.scatter(mpg_data[cols[j]], mpg_data['mpg'],  alpha=.5, color=colors[j])
    i.set_xlabel(cols[j])
    i.set_title('Pearson: %s'%mpg_data.corr().loc[cols[j]]['mpg'].round(2)+' Spearman: %s'%mpg_data.corr(method='spearman').loc[cols[j]]['mpg'].round(2))
    j+=1
```
