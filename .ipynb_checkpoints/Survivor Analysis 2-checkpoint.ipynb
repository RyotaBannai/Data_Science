```python
url_2015="http://www.pro-football-reference.com/years/2015/draft.htm"
html=urllib2.urlopen(url_2015)
strain_only_main_table=SoupStrainer(class_='table_outer_container')
s=BeautifulSoup(html,"lxml", parse_only=strain_only_main_table)
```

```python
column_headers=[th.string for th in s.find_all('tr', limit=2)[1].find_all('th')]
#column_headers= [x for x in column_headers if x is not None]
# Add the two additional column headers for the player links
column_headers.extend(["Player_NFL_Link", "Player_NCAA_Link"])
column_headers.insert(-5,'Retired_status')
column_headers.insert(-5,'Career_duration')
```

```python
table_rows=s.select("#drafts tr")[2:] #first and second row are title row
```

```python
#for elem in table_rows[0].stripped_strings: print(elem) #one row data: a player's states
```

```python
def active_year_culc(drafted_year, played_to):
    if played_to==None: return 0
    return (int(played_to)-drafted_year+1)  
```

```python
def extract_player_data(table_rows,from_):
    players_d=[]
    for player_data in table_rows:
        p=player_data
        ptn=re.compile(r'.*(\sHOF)$')
        p_columns=[td.get_text()[:-4] if ptn.match(td.get_text()) else td.string for td in p.find_all("td")]
        #p_columns=[td.get_text()[:-4] if td.get_text().endswith(" HOF") else td.get_text() for td in p.find_all("td")]
        rnd=p.find(attrs={'data-stat':'draft_round'}).string
        p_columns.insert(0,rnd)
        #Add column whether or not retire
        rt= 1 if len(p.select("td strong a"))==0 else 0
        #print(p.select("td strong a"))
        #continue
        p_columns.insert(-3,rt)
        to_RAW=p.find('td',attrs={"data-stat":"year_max"})
        to= to_RAW.string if to_RAW else None
        p_columns.insert(-3,active_year_culc(from_,to))
        if not p_columns or rnd=='Rnd': continue
        links_dict={(link.get_text()[:-4] if ptn.match(link.get_text()) else link.string) : link["href"] 
                       for link in player_data.find_all("a", href=True)}
        p_columns.append(links_dict.get(p_columns[3], ""))
        p_columns.append(links_dict.get("College Stats", ""))
        players_d.append(p_columns)
    return players_d
```

```python
data=extract_player_data(table_rows,2015)
df_2015=pd.DataFrame(data, columns=column_headers);df_2015.head(5)
```

<span style="font-size:1.2em;font-weight:bold;padding:4px;background-color:#F0FFF0"><a href="https://www.statisticshowto.datasciencecentral.com/lowess-smoothing/">What is Lowess(loess)?</a></span>

```python
# plot LOWESS curve
# set line color to be black, and scatter color to cyan
sns.regplot(x="Pick", y="CarAV", data=draft_df_2010_2, lowess=True,
            line_kws={"color": "black"},
            scatter_kws={"color": sns.color_palette()[5], "alpha": 0.5})
plt.title("Career Approximate Value by Pick")
plt.xlim(-5, 500)
plt.ylim(-5, 200)
```

```python
# Fit a LOWESS curver for each position
sns.lmplot(x="Pick", y="CarAV", data=draft_df_2010, lowess=True, hue="Pos",
           size=10, scatter=False)
plt.title("Career Approximate Value by Pick and Position")
plt.xlim(-5, 500)
plt.ylim(-1, 60)
```

```python
lm = sns.lmplot(x="Pick", y="CarAV", data=draft_df_2010_2, lowess=True, col="Pos",
                col_wrap=5, size=4, line_kws={"color": "black"},
                scatter_kws={"color": sns.color_palette()[5], "alpha": 0.7})

# add title to the plot (which is a FacetGrid)
# https://stackoverflow.com/questions/29813694/how-to-add-a-title-to-seaborn-facet-plot
plt.subplots_adjust(top=0.9)
lm.fig.suptitle("Career Approximate Value by Pick and Position",fontsize=30)

plt.xlim(-5, 500)
plt.ylim(-1, 100)
```
